```python
import cv2
import matplotlib.pyplot as plt
```

## Inference on a latin hand-writing text image

```python
import easyocr
reader_ch_e = easyocr.Reader(['ch_sim','en'])
```

```python
result1 = reader_ch_e.readtext('chinese.jpg')
img1 = cv2.imread('chinese.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
print(result1)
```

```python
for item in result1:
  if item[2] >= 0.5 :
    print(item)
```

```python
result2 = reader_ch_e.readtext('images.png')

img2 = cv2.imread('images.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
print(result2)
```
